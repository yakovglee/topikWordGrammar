```python
def replace_symbols(data, smbl):
    return data.str.replace(smbl, '/', regex=False)
```

```python
df['품사'] = replace_symbols(df['품사'], '∙')
df['어휘'] = replace_symbols(df['어휘'], '∙')

df['품사'] = replace_symbols(df['품사'], '.')
df['어휘'] = replace_symbols(df['어휘'], '.')

df['품사'] = replace_symbols(df['품사'], '・')
df['어휘'] = replace_symbols(df['어휘'], '・')

df['품사'] = replace_symbols(df['품사'], '‧')
df['어휘'] = replace_symbols(df['어휘'], '‧')

df['품사'] = replace_symbols(df['품사'], '·')
df['어휘'] = replace_symbols(df['어휘'], '·')
```

```python
df['품사'].value_counts()
```

```python
df['품사'].size
```

```python
triple_word = df[(df['품사'] == '부사/관형사/명사') | (df['품사'] == '수사/관형사/명사')]

triple_word['품사'].size
```

```python
triple_word_words = triple_word['어휘'].str.split('/', expand=True)
triple_word_values = triple_word['품사'].str.split('/', expand=True)

triple_word_words = triple_word_words.fillna('스무째00')

triple_word_words_combined = pd.concat([triple_word_words[col] for col in triple_word_words.columns], ignore_index=True)
triple_word_values_combined = pd.concat([triple_word_values[col] for col in triple_word_values.columns], ignore_index=True)


result_df_triple_word = pd.DataFrame({
    '등급': pd.concat([triple_word['등급']] * len(triple_word_words.columns), ignore_index=True),
    '어휘': triple_word_words_combined,
    '품사': triple_word_values_combined
})

combined_df = pd.concat([df, result_df_triple_word], ignore_index=True)


df = combined_df[~combined_df['품사'].isin(['부사/관형사/명사', '수사/관형사/명사'])]
```

```python
values_with_slash_words = values_with_slash['어휘'].str.split('/', expand=True)
values_with_slash_values = values_with_slash['품사'].str.split('/', expand=True)

words_combined = pd.concat([values_with_slash_words[col] for col in values_with_slash_words.columns], ignore_index=True)
values_combined = pd.concat([values_with_slash_values[col] for col in values_with_slash_values.columns], ignore_index=True)


result_df_values_with_slash = pd.DataFrame({
    '등급': pd.concat([values_with_slash['등급']] * len(values_with_slash_words.columns), ignore_index=True),
    '어휘': words_combined,
    '품사': values_combined
})

df = pd.concat([df, result_df_values_with_slash], ignore_index=True)
```
